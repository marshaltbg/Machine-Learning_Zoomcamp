```python
# Q6. Median value of horsepower
# 1 Find the median value of the horsepower column in the dataset.
median_hp = df['horsepower'].median()
print(f"Median Horsepower: {median_hp}")
```

```python
# Next, calculate the most frequent value of the same horsepower column.
mode_hp = df['horsepower'].mode()[0]  # [0] karena mode() mengembalikan series, ambil nilai pertama
print(f"Most frequent horsepower: {mode_hp}")
```

```python
# Use the fillna method to fill the missing values in the horsepower column with the most frequent value from the previous step.
df['horsepower'] = df['horsepower'].fillna(mode_hp)
```

```python
# Now, calculate the median value of horsepower once again.
new_median_hp = df['horsepower'].median()
print(f"New median HP : {new_median_hp}")
print(f"Previously Median Horsepower: {median_hp}")
```

```python
# Q7. Sum of weights
# Select only columns vehicle_weight and model_year
asia_cars_sub = asia_cars[['vehicle_weight','model_year']]
print(asia_cars_sub)
```

```python
# Select the first 7 values
asia_cars_sub_7 = asia_cars_sub.head(7)
```

```python
# Get the underlying NumPy array. Let's call it X.
X = asia_cars_sub_7.values 
```

```python
# Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.
import numpy as np
XTX = np.dot(X.T, X)
# Invert XTX.
XTX_invers = np.linalg.inv(XTX)
print(XTX_invers)
```

```python
# Create an array y with values [1100, 1300, 800, 900, 1000, 1100, 1200].
Y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
print (Y)
```

```python
# Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
W = np.dot(np.dot(XTX_invers, X.T), Y)
print (W)
```

```python
# What's the sum of all the elements of the result?
W_sum = np.sum(W)
print(W_sum)
```
